```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.patches as patches
import os
import os.path
from py_util.config import *
from py_util import config, load
import numpy as np
import csv
% matplotlib inline
import py_util.config
font = {'family' : 'serif',
        'weight' : 'normal',
        'size'   : 11,
        'sans-serif' : ['Times New Roman']
       }
rc('font', **font)
rc('legend', fontsize=7)
rc('legend', numpoints=1)
import matplotlib
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
def nums(l):
    l2 = []
    for x in l:
        if not x == '':
            l2.append(float(x))
    return np.array(l2)
cblue = '#5DA5DA'
cred = 'red'
cgreen = '#60BD68'
c4 = 'purple'
def load_thresh_stats(folder,thresh_def=''):
    dir = path_to_data + 'traces/' + folder+'/thresh_stats'+thresh_def+'.csv'
    headers = np.loadtxt(dir,dtype=bytes,delimiter=',')[0,:].astype(str)
    data_arr = np.loadtxt(dir,dtype=float,delimiter=',',skiprows=1)
    data = {}
    for i in range(len(headers)):
        data[headers[i]] = data_arr[:,i]
    firing_rows = data['ais9']>=data['thresh']
    return data, firing_rows

# global
folder = 'Jul_02_15:39:57_simple_titrate'
rc('text',usetex=False)
# import mpld3
# mpld3.enable_notebook()
    
    
def plot_collapse_sum_ais(data,firing_rows,key='intensity',val='12_'):
    collapsed_vals = {}
    data_key = data[key][firing_rows]
    data_val = data[val+'0'][firing_rows]
    for i in range(1,10):
        data_val += data[val+str(i)][firing_rows]
    return plot_collapse_simple(data_key,data_val)
def plot_collapse(data,firing_rows,key='intensity',val='16_9'):
    collapsed_vals = {}
    data_key = data[key][firing_rows]
    data_val = data[val][firing_rows]
    return plot_collapse_simple(data_key,data_val)
def plot_collapse_simple(data_key, data_val):
    collapsed_vals = {}
    for i in range(len(data_key)):
        if not data_key[i] in collapsed_vals:
            collapsed_vals[data_key[i]] = [data_val[i]]
        else:
            collapsed_vals[data_key[i]].append(data_val[i])
    ins = np.unique(data_key)
    vals,stds = [],[]
    for k in list(ins):
        vals.append(np.mean(collapsed_vals[k]))
        stds.append(np.std(collapsed_vals[k]))
    return ins,vals,stds
def mm(num_in):
    return 0.0393701*num_in
WIDTH_2 = mm(84)
```
