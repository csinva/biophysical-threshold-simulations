```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import os
import os.path
from py_util.config import *
from py_util import config, load
import numpy as np
import csv
% matplotlib inline
import py_util.config
font = {'family' : 'sans-serif',
        'weight' : 'normal',
        'size'   : 11,
        'sans-serif' : ['Helvetica']
       }
rc('font', **font)
rc('text',usetex=False)
rc('legend', fontsize=7)
rc('legend', numpoints=1)
import matplotlib
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
def nums(l):
    l2 = []
    for x in l:
        if not x == '':
            l2.append(float(x))
    return np.array(l2)
cblue = '#5DA5DA'
cred = 'red'
cgreen = '#60BD68'
def load_thresh_stats(folder,thresh_def=''):
    dir = path_to_data + 'traces/' + folder+'/thresh_stats'+thresh_def+'.csv'
    headers = np.loadtxt(dir,dtype=bytes,delimiter=',')[0,:].astype(str)
    data_arr = np.loadtxt(dir,dtype=float,delimiter=',',skiprows=1)
    data = {}
    for i in range(len(headers)):
        data[headers[i]] = data_arr[:,i]
    firing_rows = data['ais9']>=data['thresh']
    return data, firing_rows
# global
folder = 'Jul_02_15:39:57_simple_titrate'
rc('text',usetex=False)
# import mpld3
# mpld3.enable_notebook()
def plot_collapse(data,firing_rows,key='intensity',val='16_9'):
    collapsed_vals = {}
    data_key = data[key][firing_rows]
    data_val = data[val][firing_rows]
    return plot_collapse_simple(data_key,data_val)
def plot_collapse_simple(data_key, data_val):
    collapsed_vals = {}
    for i in range(len(data_key)):
        if not data_key[i] in collapsed_vals:
            collapsed_vals[data_key[i]] = [data_val[i]]
        else:
            collapsed_vals[data_key[i]].append(data_val[i])
    ins = np.unique(data_key)
    vals,stds = [],[]
    for k in list(ins):
        vals.append(np.mean(collapsed_vals[k]))
        stds.append(np.std(collapsed_vals[k]))
    return ins,vals,stds
```

```python
folder = 'Jul_02_15:39:57_simple_titrate'
data, firing_rows = load_thresh_stats(folder)
fig = plt.figure(figsize=(6, 2.5), dpi=600, tight_layout=True, facecolor='white')  # should be 600
plt.subplot(121)
plt.title('A',loc='left',size=11,fontweight='bold')
ins,vals,stds = plot_collapse_simple(data['intensity'][firing_rows],data['t'][firing_rows]-50)
plt.plot(ins,vals,'o',markeredgewidth=0,color=cblue,label='Nav 1.6')
plt.errorbar(ins,vals,yerr=stds,color='black',linestyle='None')

# plt.plot(data['intensity'],data['t']-50,'o',markeredgewidth=0,color=cblue)
plt.xlabel('Stimulus Intensity (nA)')
plt.ylabel('Time to Threshold (ms)')
plt.subplot(122)
plt.title('B',loc='left',size=11,fontweight='bold')
plt.plot(data['intensity'],data['thresh'],'o',markeredgewidth=0,color=cblue)
plt.xlabel('Stimulus intensity (nA)')
plt.ylabel('Threshold (mV)')
plt.ylim((-54,-51))
plt.savefig(path_to_folder + 'py/plots/thresh_v_lambda.pdf')
plt.show()
```
